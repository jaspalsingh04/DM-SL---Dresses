```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv("C://Users//Jayana//Desktop//dresses.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
df = df.fillna(df.mode().iloc[0])
df
```

```python
df.isnull().sum()
```

```python
df["Season_en"]= le.fit_transform(df["Season"])
df
```

```python
df["Season_en"]= le.fit_transform(df["Season"])
df["Style_en"]= le.fit_transform(df["Style"])
df["Neckline_en"]= le.fit_transform(df["Neckline"])
df["Waseline_en"]= le.fit_transform(df["Waseline"])
df["Material_en"]= le.fit_transform(df["Material"])
df["Fabric_en"]= le.fit_transform(df["Fabric"])
df["Decoration_en"]= le.fit_transform(df["Decoration"])
df["Pattern_en"]= le.fit_transform(df["Pattern"])
df
```

```python
y= df.Class
```

```python
from sklearn.model_selection import train_test_split
```

```python
dropped_df = df.drop(['Style','Recommendation','Season','Neckline','Sleeve_length','Waseline','Material','Fabric','Decoration','Pattern'], axis= 1)
```

```python
X_train,X_test, y_train, y_test = train_test_split(dropped_df, y, test_size=0.2)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression(class_weight= 'balanced')
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cr= confusion_matrix(y_test, y_predicted)
cr
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cr, annot= True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted)
```

```python
from sklearn.metrics import precision_score
pre= precision_score(y_test, y_predicted)
pre
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
f1=f1_score(y_test, y_predicted)
recall= recall_score(y_test, y_predicted)

recall
```
